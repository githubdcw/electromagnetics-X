Find electric field on x-axis produced by a circular line charge of $\rho_l$ C/m centered at the orgin with the radius of 'a' m.

สมการหาสนามไฟฟ้าจากประจุแบบจุด้วยกฎของคูลอมบ์  

$$
d\mathbf E(x) = \frac{dQ}{4\pi \varepsilon_o} \frac{\mathbf {r-r'}}{|\mathbf {r-r'}|^3}
$$  

เมื่อ  

$$
dQ = \rho_l a d \phi'\\
\mathbf r = x \mathbf a_x\\
\mathbf r' = a \mathbf a'_\rho \\
\mathbf a'_\rho = \cos \phi' \mathbf a_x + \sin \phi' \mathbf a_y
$$

แทนค่า $dQ$ , $\mathbf r$ และ $\mathbf r'$ ในสมการกฎของคูลอมบ์จะได้สมการด้านล่าง    

$$
d\mathbf E(x) = \frac{\rho_l a}{4\pi \varepsilon_o} \frac{x\mathbf {a}_x-a \mathbf{a}'_\rho}{|x\mathbf {a}_x-a \mathbf{a}'_\rho|^3} d \phi'\\
d\mathbf E(x) = \frac{\rho_l a}{4\pi \varepsilon_o} \frac{(x-a \cos(\phi'))\mathbf {a}_x-a \sin(\phi')  \mathbf{a}_y}{|(x-a \cos(\phi'))^2+(a \sin(\phi'))^2|^\frac{3}{2}} d \phi'
$$   


อินทิเกรตส่วนที่คูณกับ $\mathbf a_x$  

$$
\int_{\phi'=0}^{\phi'=2 \pi} \frac{(x-a \cos(\phi'))}{|(x-a \cos(\phi'))^2+(a \sin(\phi'))^2|^\frac{3}{2}} d \phi'
$$

ต้องกำหนดค่า $a$ กับ $x$ เพื่อให้สามารถใช้วิธีอินทิเกรตเชิงตัวเลขได้ เช่นให้ รัศมี $a=1$ และ ตำแหน่ง $x=2$

```python
#Integrate ด้วยการแบ่งปัญหาเป็น 10 ชิ้น คำนวนทีละชิ้นแล้วนำมาบวกกัน
import numpy as np
N = 10
delta = 2*np.pi/N
phii = delta/2 + delta*np.array(range(N))
cosphii = np.cos(phii)
sinphii = np.sin(phii)
x = 2
a = 1
tmp = delta*(x-a*cosphii)/((x-a*cosphii)**2+(a*sinphii)**2)**(3/2)
print(sum(tmp))
```

```python
#plot the value of each piece
import matplotlib.pyplot as plt
plt.plot(phii,tmp)
plt.grid()
plt.show()
```

อินทิเกรตส่วนที่คูณกับ $\mathbf a_y$  

$$
\int_{\phi'=0}^{\phi'=2 \pi} \frac{a \sin(\phi')}{|(x-a \cos(\phi')^2+(a \sin(\phi'))^2|^\frac{3}{2}} d \phi'
$$

ถ้าพิจารณาจากความสมมาตร ด้าน $y>0$ มีประจุเท่ากับด้าน $y<0$   เราจะสามารถคาดการณ์ได้ว่าสนามไฟฟ้าทิศทาง $\mathbf a_y$ จะหักล้างกันหมดเป็นศูนย์โดยไม่ต้องทำการอินทิเกรต  
ถ้าใช้วิธีอินทิเกรตเชิงตัวเลข จะได้ค่าที่น้อยมากๆ  
ถ้าใช้ https://www.integral-calculator.com/ จะสามารถหา closed form ของผลลัพท์ได้แต่อาจยากเกินไปเพราะต้องให้ $u=  (x-a \cos(\phi')^2+(a \sin(\phi'))^2$.  
ถ้าหาผลลัพท์ด้วยวิธีเชิงตัวเลขจะง่ายกว่า  

```python
#Integrate ด้วยการแบ่งปัญหาเป็น 50 ชิ้น คำนวนทีละชิ้นแล้วนำมาบวกกัน
import numpy as np
N = 50
delta = 2*np.pi/N
phii = delta/2 + delta*np.array(range(N))
cosphii = np.cos(phii)
sinphii = np.sin(phii)
x = 2
a = 1
tmp = delta*sinphii/((x-a*cosphii)**2+(a*sinphii)**2)**(3/2)
print(sum(tmp))
```

```python
import matplotlib.pyplot as plt
plt.plot(phii,tmp)
plt.grid()
plt.show()
```
